```
# -*- coding: utf-8 -*-
"""
Created on Thu Aug 15 15:41:39 2019

@author: srg
"""

import numpy as np
import tensorflow as tf
import pandas as pd

train = []
train_out = []

for i in range(10):
	for j in range(10):
		train.append([str(i),str(j)])
		train_out.append([str((i+j)//10), str((i+j)%10)])
		
train = np.array(train)
train_out = np.array(train_out)

train = pd.DataFrame(train, columns = ['tens', 'units'])
train = pd.get_dummies(train)
#train = train.head(10)

train_out = pd.DataFrame(train_out, columns = ['tens', 'units'])
train_out = pd.get_dummies(train_out)
#train_out = train_out.head(10)
#print(train)
#
#train = np.reshape(train, (100,20,1))
#train_out = np.reshape(train_out, (100,12,1))
test = train
test_out = train_out



print(train.shape)
print(train_out.shape)


def calculator():
	
	model = tf.keras.Sequential([
	
	tf.keras.layers.Dense(20, input_shape=(20,)),
	tf.keras.layers.Dense(240, activation=tf.nn.relu),
#tf.keras.layers.Dense(10, activation=tf.nn.relu),
 #   tf.keras.layers.Dense(40, activation=tf.nn.relu),
	tf.keras.layers.Dense(12, activation=tf.nn.sigmoid)
	
	])
	
	model.compile(optimizer='adam',
			  loss = tf.keras.losses.MeanAbsoluteError(),
			  metrics = ['accuracy'])
	#print(model.summary())
	
	return model
print("here")
calc = calculator()
print("fitting")
acc = True
cnt = 600#650
while(acc and cnt<100000):
	calc = calculator()
	calc.fit(train, train_out, epochs = cnt, verbose = 0)
	print(cnt)
	test_loss, test_acc = calc.evaluate(test, test_out)
	#print("losses = " , test_loss, test_acc)
	prediction = calc.predict(test)
	predict = []
	prediction = prediction.tolist()
	for j in prediction:
		predict.append([])
		for i in range(12):
			if j[i]==max(j[0],j[1]) and i<2 :
				predict[-1].append(1)
			elif j[i]==max(j[2:12]) and i>=2:
				predict[-1].append(1)
			else:
				 predict[-1].append(0)
		#print(predict)
	xx = np.array(train_out).tolist()
	acc = not (xx == predict)
	cnt += 1
print(*predict, sep = '\n')
print(cnt)
calc.save('model'+str(cnt))
#test_out = [[int(i[0][0]*10),int(i[1][0]*10)] for i in test_out.tolist()]
```
